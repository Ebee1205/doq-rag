```python
from google.colab import drive
from pathlib import Path

drive.mount("/content/drive")

# 작업 폴더
WORK_DIR = Path("/content/drive/MyDrive/작업/앙코코-도큐/RAG")

# GitHub 저장소가 위치할 폴더
REPO_DIR = WORK_DIR / "doq-rag"

# GitHub 저장소 정보
GITHUB_USER = "Ebee1205"
REPOSITORY = "doq-rag"
REPO_URL = f"https://github.com/{GITHUB_USER}/{REPOSITORY}.git"

WORK_DIR.mkdir(parents=True, exist_ok=True)

print("작업 폴더:", WORK_DIR)
print("저장소 폴더:", REPO_DIR)
```

```python
import os
from google.colab import userdata

GITHUB_TOKEN = userdata.get("GITHUB_TOKEN")

if not GITHUB_TOKEN:
    raise ValueError(
        "Colab 보안 비밀에 GITHUB_TOKEN이 없습니다.\n"
        "왼쪽 열쇠 아이콘에서 토큰을 등록해주세요."
    )

# 이후 실행되는 git 프로세스가 사용할 환경변수
os.environ["GITHUB_TOKEN"] = GITHUB_TOKEN

print("GitHub 토큰을 불러왔습니다.")
```

```python
import subprocess

def run_git(args, cwd=None, check=True):
    result = subprocess.run(
        ["git", *args],
        cwd=cwd,
        text=True,
        capture_output=True,
    )

    if result.stdout.strip():
        print(result.stdout.strip())

    if result.stderr.strip():
        print(result.stderr.strip())

    if check and result.returncode != 0:
        raise RuntimeError(
            f"Git 명령 실행 실패: git {' '.join(args)}"
        )

    return result


credential_helper = (
    "!f() { "
    "echo username=x-access-token; "
    "echo password=$GITHUB_TOKEN; "
    "}; f"
)

run_git([
    "config",
    "--global",
    "credential.helper",
    credential_helper,
])

print("GitHub 인증 설정 완료")
```

```python
gitignore_path = REPO_DIR / ".gitignore"

gitignore_content = """\
.ipynb_checkpoints/
__pycache__/
*.pyc
.env
.DS_Store
"""

gitignore_path.write_text(
    gitignore_content,
    encoding="utf-8",
)

print(gitignore_path.read_text(encoding="utf-8"))
```
